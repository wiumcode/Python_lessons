# Popularidade em um site

```python
import requests

url = "https://sicnoticias.pt/"

result = requests.get(url)

texto = result.text

palavra1 = input()
palavra2 = input()

qtd_palavra1 = texto.count(palavra1)
qtd_palavra2 = texto.count(palavra2)

print(palavra1, qtd_palavra1, "x", palavra2, qtd_palavra2)
```

```python
import requests

lista_urls = ["https://sicnoticias.pt/", "https://cnnportugal.iol.pt/", "https://www.rtp.pt/", "https://tvi.iol.pt/"]

palavra1 = input()
palavra2 = input()
qtd_palavra1=0
qtd_palavra2=0

for url in lista_urls:
    
    result = requests.get(url)
    texto = result.text

    qtd_palavra1 += texto.lower().count(palavra1)
    qtd_palavra2 += texto.lower().count(palavra2)

print(palavra1, qtd_palavra1, "x", palavra2, qtd_palavra2)
```

# Contando homens e mulheres nos filmes

```python
len(dict_people["results"])
```

```python
qtd_homens = 0
qtd_mulheres = 0
qtd_na = 0

for numero in range(10):
    sexo = dict_people["results"][numero]["gender"]
    if sexo == "male":
        qtd_homens+=1
    elif sexo == "female":
        qtd_mulheres+=1
    else:
        qtd_na+=1
        

print("Quantidade de homens: {}, Quantidade de mulheres: {}, Quantidade de n/a: {}".format(qtd_homens, qtd_mulheres, qtd_na))
        
```

# Mudando url

```python
%%time
import requests

url = "https://swapi.dev/api/people/"

qtd_homens = 0
qtd_mulheres = 0
qtd_na = 0
        
while True:
    print("Analizando a pagina", url)
    try:
        result = requests.get(url)

        dict_people = result.json()

        for numero in range(10):
            sexo = dict_people["results"][numero]["gender"]
            if sexo == "male":
                qtd_homens+=1
            elif sexo == "female":
                qtd_mulheres+=1
            else:
                qtd_na+=1

        proxima_pagina = dict_people["next"]

        url = proxima_pagina
    
    except:
        break
        
print("Quantidade de homens: {}, Quantidade de mulheres: {}, Quantidade de n/a: {}".format(qtd_homens, qtd_mulheres, qtd_na))
```
