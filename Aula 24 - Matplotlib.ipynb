# Teoria - Justificação de usar Matplotlib em Pandas

O matplotlib foi criado para ser uma biblioteca relacionada com numpy

- numpy = vetores(array) / listas
- pandas = tabelas / DataFrames

```python
pandas.Series (coluna) <<<< numpy.Lista <<<< Matplotlib
```

# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Exemplo 1 - Seno

```python
x = np.arange(0, 10, 0.1)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
plt.title("Grafico de Seno")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
```

```python
idades = [10, 18, 19, 20, 50]
gastos = [2, 40, 30, 15, 100]

plt.plot(idades, gastos)
plt.title("Gastos no Arraial por idade")
plt.xlabel("Idades")
plt.ylabel("Gastos")
plt.show()
```

```python
df_arraial = pd.DataFrame()
df_arraial["idades"] = [10, 18, 19, 20, 50]
df_arraial["gastos"] = [2, 40, 30, 15, 100]

plt.plot(df_arraial.idades, df_arraial.gastos)
plt.title("Gastos no Arraial por idade")
plt.xlabel("Idades")
plt.ylabel("Gastos")
plt.show()
```

# Exemplo 2 - Histograma

```python
preco = [10, 18, 19, 20, 50, 88, 67, 11, 20, 34, 7, 65, 45]

plt.hist(preco,  bins=2)
plt.title("Histograma dos Precos dos Produtos")
plt.xlabel("Idades")
plt.ylabel("Frequencia")
plt.show()
```

# Exemplo - Linha

```python
linha = np.arange(0,10,0.2)
```

```python
linha
```

```python
plt.plot(linha)
```

```python
seno = np.sin(linha)

plt.plot(seno)
plt.show()
```

```python
plt.plot(seno)
plt.plot(seno*-1)
plt.show()
```

## alterando tamanho

```python
figure = plt.figure(figsize=(20,10))
plt.title("Função Seno")
plt.plot(seno)
plt.plot(seno*-1)
plt.show()
```

```python
figure.savefig("teste_seno.png")
```

## mudando cor, pontos e font

```python
figure = plt.figure(figsize=(20,10))
plt.title("Função Seno", fontsize=20)
plt.plot(seno, color="green", marker='o')
plt.plot(seno*-1, 'r--')
plt.show()
```
