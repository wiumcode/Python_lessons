# Teoria - Justificação de usar Matplotlib em Pandas

O matplotlib foi criado para ser uma biblioteca relacionada com numpy

- numpy = vetores(array) / listas
- pandas = tabelas / DataFrames

pandas.Series (coluna) <<<< numpy.Lista <<<< Matplotlib

# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Gráfico com coordenadas X e Y

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) 

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3)
```

# Outros tipos de gráficos

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

x = np.arange(0, 6, 1)

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

# Anotações

```python
fig, ax = plt.subplots()

ax.plot(xx, xx**2, label="formula1")
ax.plot(xx, xx**3, label="formula2")

ax.text(0.15, 0.2, "y-x^2", fontsize=20, color = "blue")
ax.text(0.65, 0.1, r"$y-x^3$", fontsize=20, color = "orange")

ax.legend()
```

# Colormap e contorno

```python
def f(x,y): 
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

#contourf desenha o preenchimento contourf(x,y,funcao,niveis)
#contour desenha as linhas contourf(x,y,funcao,niveis)
plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X,Y), 8, colors='black')
plt.show()
```

# Gráficos 3d

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
plt.figure(figsize=(20,5))
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)

fig = plt.figure()
ax = Axes3D(fig)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

plt.show()
```
