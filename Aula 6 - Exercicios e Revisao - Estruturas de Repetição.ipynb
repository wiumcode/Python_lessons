# While

while condicao:

    codigo aqui vai repetir

```python
senha = "123"

password = input("Digite sua senha: ")

while senha != password:
    print("Senha esta errada")
    password = input("Tente novamente: ")

print("Bem vindo!")
```

```python
lista_compras = []
item = input("Insira um item na sua lista de compras ('fim' para terminar) : ")

while item in ["fim"]:
    lista_compras.append(item)
    item = input("Insira um item na sua lista de compras ('fim' para terminar) : ")
    
print(lista_compras)
```

```python
estudante_professor = input("Vc eh estudante ou professor? (S/N) :").upper()

while estudante_professor not in ["S", "N"]:
    estudante_professor = input("Vc eh estudante ou professor? (S/N) :")
    
if estudante_professor == "S":
    print("Legal")
else:
    print("Vai estudar!")
```

```python
senha = "123"

for chances in range(3):
    password = input("Digite sua senha: ")
    if senha == password:
        print("Bem vindo!")
        break
    else:
        print("Senha errada!")
```

```python
senha = "123"

password = input("Digite sua senha: ")

tentativas = 0

while senha != password:
    tentativas +=1
    print("Senha esta errada")
    if tentativas >2:
        break
    password = input("Tente novamente: ")
    
    
if tentativas < 3:
    print("Bem vindo!")
else:
    print("Acesso bloqueado")
```

```python
x = 10
while True:
    x *=2
    if x > 100:
        break

print(x)
```

```python
password = "bea"

password_entry = input("type password")

tries = 0

while password != password_entry:
    tries += 1
    print("wrong password")
    if tries > 2:
        break
    password_entry = input("try again")
    
if tries < 3:
    print("Welcome!")
else:
    print("Access denied")
```
