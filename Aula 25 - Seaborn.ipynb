# Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

# Gráficos de Distribuição

## 1. Histograma

```python
data = np.random.randn(1000) # Dados aleatórios
```

```python
data
```

```python
sns.distplot(data, kde=False)
plt.title("Histograma")
plt.xlabel("Valores")
plt.ylabel("Frequencia")
plt.show()
```

## 2. Grafico de densidade

```python
sns.kdeplot(data, shade=True)
plt.title("Gráfico de densidade")
plt.xlabel("Valores")
plt.ylabel("Densidade")
plt.show()
```

# Gráficos de Relacionamentos

## 1. Scatter plot

```python
altura = np.random.randint(150, 190, size=100)
peso = altura + np.random.randint(-10, 10, size=100)
```

```python
sns.scatterplot(x=altura, y=peso)
plt.title("Altura versus Peso")
plt.xlabel("Altura")
plt.ylabel("Peso")
plt.show()
```

## 2. Pair plot

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.shape
```

```python
titanic.head()
```

```python
sns.pairplot(iris, hue="species")
plt.title("Pair plot - Iris dataset")
plt.show()
```

```python
sns.pairplot(titanic[["sex", "age", "survived"]], hue="sex")
plt.title("Pair plot - Titanic dataset")
plt.show()
```

## 3. Heatmap

```python
data = np.random.rand(10,10)
```

```python
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Heatmap - Matriz de correlação")
plt.show()
```

# Grágicos Categóricos

## 1. Bar plot

```python
# Exemplo: Contagem de valores categóricos
data = ['A', 'A', 'B', 'B', 'B', 'C']

sns.countplot(data) # Cria um bar plot de contagem
plt.title('Count plot - Contagem de valores categóricos')
plt.xlabel('Valores')
plt.ylabel('Contagem')
plt.show()
```

## 2. Box plot

```python
sns.boxplot(x=titanic["age"])
```
